```python
import numpy as np
from matplotlib import pyplot as plt

COLORS = np.array([[142/255, 1.0, 0.0], [97/255, 0.0, 1.0],
                  [1.0, 0.0, 89.0/255], [1.0, 220/255, 0.0],[222/255., 1.0, 0.0],
                  [22/255., 33.0/255., 0.0]])
```

```python
def f(z):
    return z**3 - 1


def fp(z):
    return 3*(z**2)
```

```python
root1, root2, root3 = np.roots([1,0,0,-1])
```

```python
xmin = -2
xmax = 2
ymin = -2
ymax = 2
nx = 200
ny = 200
X = np.linspace(xmin, xmax, nx)
Y = np.linspace(ymin, ymax, ny)
```

```python
x,y=np.meshgrid([1,2,3,4], [1,2,3,4])
print(x[3],y[3])
plt.scatter(x,y)
```

```python
X, Y = np.meshgrid(X, Y)
```

```python
Z = X + 1j*Y
```

```python
for i in range(40):
    Z = Z - f(Z)/fp(Z)
```

```python
eps = 0.001

Z1 = abs(Z-root1) < eps
Z2 = abs(Z-root2) < eps
Z3 = abs(Z-root3) < eps
```

```python
arr = np.zeros(shape=(200, 200, 3))
for i in range(nx):
    for j in range(ny):
        if Z1[i][j]:
            arr[i][j] = COLORS[0]
        elif Z2[i][j]:
            arr[i][j] = COLORS[1]
        elif Z3[i][j]:
            arr[i][j] = COLORS[2]

print(arr.shape)

plt.imshow(arr)

plt.show()
```

## Let's compute some more fractals and I'll show an easier way to work with this..
JUST COPY PASTE FROM ABOVE

```python
xmin = -2
xmax = 2
ymin = -2
ymax = 2
nx = 250
ny = 250
X = np.linspace(xmin, xmax, nx)
Y = np.linspace(ymin, ymax, ny)
X,Y=np.meshgrid(X,Y)

Z = X + 1j*Y
for i in range(50):
    Z = Z - p(Z)/p.deriv()(Z)
```

```python
eps = 0.001
Z_T = []
for r in p.roots:
    Z_T.append(abs(Z-r) < eps)
```

```python
# Z_T = []
colored_arr = np.zeros(shape=(nx, ny, 3))

for i in range(nx):
        for j in range(ny):
            for k, z_t in enumerate(Z_T):
                if z_t[i][j]:
                    colored_arr[i][j] = COLORS[k]
                    # colored_arr[i][j] = get_color()
```

```python
plt.figure(figsize=(15,15))
plt.imshow(colored_arr)
```
